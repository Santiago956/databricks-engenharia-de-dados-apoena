```sql
SELECT 
    c_mktsegment,
    COUNT(c_custkey) AS customer_count
FROM samples.tpch.customer
GROUP BY c_mktsegment
HAVING customer_count > 150000
ORDER BY customer_count DESC;
```

```sql
SELECT c_custkey, COUNT(c_custkey) AS occurrences
FROM samples.tpch.customer
GROUP BY c_custkey
HAVING COUNT(*) > 1;
```
